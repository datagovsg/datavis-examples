```python
import pandas as pd
```

# Load the original csv
Notice that the data provided is NOT in a [tidy data format](http://vita.had.co.nz/papers/tidy-data.html).  


```python
df = pd.read_csv('../data/burtin.csv')
df
```

# Convert to tidy format
This is a narrow format for tidy data.

```python
tidy = pd.melt(df,
               id_vars=['Bacteria', 'Gram Staining'], 
               value_vars=['Penicilin', 'Streptomycin', 'Neomycin'],
               var_name='Antibiotic',
               value_name='MIC')
tidy.head()
```

```python
# save you data
tidy.to_csv('../data/burtin-tidy.csv', index=False)
```
